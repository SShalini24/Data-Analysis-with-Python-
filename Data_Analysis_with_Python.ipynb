**Data Cleaning Replace ("--" to "NaN")**

```python
data = data.replace("--",np.NaN,regex=True)
```

**Checking the Null Values**

```python
data.isnull().sum()
```

```python
per_missing = data.isnull().sum()*100/len(data)
per_missing
```

```python
sns.heatmap(data.isnull())
```

```python
sns.heatmap(data.isnull())
```

```python
data.dropna(axis=0,inplace=True)   # Deleting the missing rows
```

```python
data.head()
```

**Data Cleaning (Rank Column)**

```python
data.dtypes
```

```python
data['Rank'] = data['Rank'].str[0:-2]
```

```python
data
```

```python
data['Rank']=data['Rank'].str.replace(',' ,'').astype('int')
```

```python
data['Rank']
```

```python
data.dtypes
```

```python
data
```

**Data Cleaning (Video uploads and Subscriber)**

```python
data.head()
```

```python
data['Video Uploads'] = data['Video Uploads'].astype('int')
```

```python
data['Subscribers'] = data['Subscribers'].astype('int')
```

```python
data.dtypes
```

**Data Cleaning [Grade Column]**

```python
data['Grade'].unique()
```

```python
data['Grade']=data['Grade'].map({'A++ ':5,'A+ ':4,'A ':3,'A- ':2,'B+ ':1})
```

```python
data['Grade'].unique()
```

```python
data.head()
```

**Finda Average Views for Each Channel**

```python
data.columns
```

```python
data['Avg_views'] = data['Video views']/data['Video Uploads']
```

```python
data.head()
```

**Top Five Channels with maximum number of video uploads**

```python
data.sort_values(by='Video Uploads',ascending=False).head()
```
